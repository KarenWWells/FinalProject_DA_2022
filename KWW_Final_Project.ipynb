```python
#Importing data analysis tools
import pandas as pd
import seaborn as sns

# Retrieve Data from Github
df = pd.read_csv('https://raw.githubusercontent.com/KarenWWells/FinalProject_DA_2022/main/Pocket%20Measurements.csv')

#Print Column Titles to Identify names of Columns
print(df.columns)
```

```python
#Pie Chart to Examine Distribution of brands
#Found discrepancies in report of same brands with different spelling and fixed

from matplotlib import pyplot as plt

brands = df['brand'].value_counts()
brands.plot(kind='pie')

plt.title('Distribution of Brands of Jeans')
```

```python
# Creating histograms to explore the distribution of data for the various parameters
df[['maxHeightBack', 'minHeightBack','maxWidthBack',
       'minWidthBack','maxHeightFront', 'minHeightFront','maxWidthFront',
       'minWidthFront','price']].hist(figsize=(14,9))
plt.show()
```

```python
#Created bpxplots to look for outliers
#All measurements appear on the same scale
df[['maxHeightBack', 'minHeightBack','maxWidthBack',
       'minWidthBack','maxHeightFront', 'minHeightFront','maxWidthFront',
       'minWidthFront']].boxplot(figsize=(14,9))
```

```python
#Utilizing dfstatsum, measurment summary statistics for ALL jeans are provided
dfstatsum = df[['maxHeightBack', 'minHeightBack','maxWidthBack',
       'minWidthBack','maxHeightFront', 'minHeightFront','maxWidthFront',
       'minWidthFront','price']]
mean = dfstatsum.mean()
median = dfstatsum.median()
StDev = dfstatsum.std()
keys = ["Mean", "Median", "St Dev"]
table = pd.concat([mean, median, StDev], axis = 1, keys = keys)

print(table)
```

```python
#Using df.describe() to create summary statistics men's jeans
from matplotlib import pyplot as plt
dfmen = df[df['menWomen'] == 'men']

dfmen.describe()
```

```python
#Using df.describe() to create summary statistics women's jeans
from matplotlib import pyplot as plt
dfwomen = df[df['menWomen'] == 'women']

dfwomen.describe()
```

```python
#Boxplots with Front Dimensions Separated by Men and Women
plt.figure(1, figsize=(30,50))

for i in range (6,11):
  plt.subplot(6,10,i)
  bpm = plt.boxplot(dfmen[df.columns[i]], patch_artist=True)
  bpw = plt.boxplot(dfwomen[df.columns[i]], patch_artist=True)
  plt.title(df.columns[i])
  #plt.legend([bpw["boxes"][0], bpm["boxes"][0]], ['Women', 'Men'])

  for box in bpm['boxes']:
    box.set(facecolor = 'blue')
    plt.legend([bpm["boxes"][0], bpw["boxes"][0]], ['Men', 'Women'])

  for box in bpw['boxes']:
    box.set(facecolor = 'pink')
    plt.legend([bpm["boxes"][0], bpw["boxes"][0]], ['Men', 'Women'])

plt.show()
```

```python
#Create a pie chart to examine the distribution of "Men" data versus "Women"
from matplotlib import pyplot as plt

brands = df['menWomen'].value_counts()
brands.plot(kind='pie')

plt.title('Distribution of Samples for Men and Women')
```

```python
#Creating a scatterplot to examine Maximum
#Front Pocket Dimesions for Men and Women vs Price
from matplotlib import pyplot as plt
dfmen = df[df['menWomen'] == 'men']
dfwomen = df[df['menWomen'] == 'women']

mprice = dfmen['price']
wprice = dfwomen['price']
mmhf = dfmen['maxHeightFront']
wmhf = dfwomen['maxHeightFront']

plt.scatter(mprice, mmhf, label="Men's Jeans", color='blue')
plt.scatter(wprice, wmhf, label="Women's Jeans", color='pink')
plt.title('Maximum Front Pocket Height versus Price')
plt.xlabel('Price')
plt.ylabel('Maximum Front Pocket Height')
plt.show()
```

```python
#Men's Jeans and Price scatterplot with Linear Regression
#Maximum Front Pocket Height
#Outlier analysis
from matplotlib import pyplot as plt
dfmen = df[df['menWomen'] == 'men']

mprice = dfmen['price']

mmhf = dfmen['maxHeightFront']

plt.title("Men's Maximum Front Pocket Height versus Price")

sns.regplot(x= mprice, y = mmhf)
```

```python
#Women's Jeans and Price scatterplot with Linear Regression
#Maximum Front Pocket Height
#Outlier Analysis
from matplotlib import pyplot as plt

dfwomen = df[df['menWomen'] == 'women']
wprice = dfwomen['price']
wmhf = dfwomen['maxHeightFront']

plt.title("Women's Maximum Front Pocket Height versus Price")

sns.regplot(x= wprice, y = wmhf)
```

```python
#Examine the relationship between Front and Back Max Pocket Height
from matplotlib import pyplot as plt
dfmen = df[df['menWomen'] == 'men']
dfwomen = df[df['menWomen'] == 'women']


mmhb = dfmen['maxHeightBack']
wmhb = dfwomen['maxHeightBack']
mmhf = dfmen['maxHeightFront']
wmhf = dfwomen['maxHeightFront']

plt.scatter(mmhb, mmhf, label="Men's Jeans", color='blue')
plt.scatter(wmhb, wmhf, label="Women's Jeans", color='pink')
plt.title('Maximum Front Pocket Height versus Max Back Pocket Height')
plt.xlabel('Maximum Back Pocket Height')
plt.ylabel('Maximum Front Pocket Height')

plt.show()
```
